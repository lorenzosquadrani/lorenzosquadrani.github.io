```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Persistent sodium current $I_{NaP}$


$
 I_{\text{NaP}} = g_{\text{NaP}} p_\infty (V) (V - V_{\text{Na}})
$

$
p_\infty (V) = \frac{1}{1 +  e^{-\frac{V - \theta_{\text{NaP}}}{\sigma_{\text{NaP}}}}}
$

```python
def p_infty(x):
    theta = -50
    sigma = 5
    return 1 / (1 + np.exp(-(x - theta) / sigma))

x = np.linspace(-100, 50, 1000)
y = p_infty(x)

plt.plot(x, y)
plt.title("Plot of p_infty")
plt.xlabel("x")
plt.ylabel("p_infty(x)")
plt.grid()
plt.show()
```
