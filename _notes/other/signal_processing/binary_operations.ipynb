```python
import numpy as np
import matplotlib.pyplot as plt
import skimage.morphology as skimo
```

```python
np.random.seed(3)
x = np.random.standard_normal(20)>-0.1
opened_x = skimo.area_opening(x, area_threshold=3, connectivity=1)

plt.figure(figsize=(10,1))
plt.axis('off')
plt.bar(range(20), x)

plt.figure(figsize=(10,1))
plt.axis('off')
plt.bar(range(20), opened_x)

plt.show()
```

```python
from scipy.ndimage import binary_opening

np.random.seed(3)
x = np.random.standard_normal((20, 20))>0
opened_x = binary_opening(x, structure=np.ones((3,1)))

fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(x, cmap='gray')
ax[0].axis('off')
ax[1].imshow(opened_x, cmap='gray')
ax[1].axis('off')

plt.show()
```
