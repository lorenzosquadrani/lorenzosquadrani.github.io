```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
seed = 32
rng = np.random.default_rng(seed)
```

```python
def generate_signal(t, params):
    return t

def generate_noise(t, params):

    dist = params['dist']

    if dist=='gaussian':
        assert {'mu', 'sigma'} <= set(params), 'Gaussian distribution requires the parameters mu and sigma!' 
        return rng.normal(params['mu'], params['sigma'], len(t))
```

```python
t = np.linspace(0, 10, 1000)

signal = generate_signal(t, {})
noise = generate_noise(t, {'dist':'gaussian', 'mu': 0., 'sigma': 1})

plt.plot(signal * noise)
plt.plot(signal)
```
