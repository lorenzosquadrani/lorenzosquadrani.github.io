```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import diplib as dip
import cv2
```

```python
image = np.random.rand(1000, 1000)
plt.imshow(image, cmap='gray')
plt.show()
```

```python
sigmas = [1, 5, 15, 30]

fig, axs = plt.subplots(1, 4, figsize=(20, 5))

for ax, sigma in zip(axs,sigmas):

    methods = ['scipy', 'DipLib', 'OpenCV']
    result_scipy = %timeit -o scipy.ndimage.gaussian_filter(image, sigma=sigma)
    result_dip = %timeit -o dip.Gauss(image, sigmas=sigma)
    result_cv2 = %timeit -o cv2.GaussianBlur(image, (0, 0), sigma)
    results = [result_scipy, result_dip, result_cv2]


    for result, method in zip(results, methods):
        mean = result.average
        std = result.stdev
        ax.bar(method, mean, yerr=std, capsize=5)

plt.show()
```
